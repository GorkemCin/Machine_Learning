## **1. üå≥ Introduction** / _Giri≈ü_

## üå≥ What is K-Means? / K-Means Nedir?

**K-Means**, etiketlenmemi≈ü (unlabeled) veriler √ºzerinde √ßalƒ±≈üan bir **unsupervised learning (denetimsiz √∂ƒürenme)** algoritmasƒ±dƒ±r.  
Ama√ß, veri noktalarƒ±nƒ± **benzerliklerine g√∂re K adet k√ºmeye (cluster)** ayƒ±rmaktƒ±r.

Algoritma, her k√ºme i√ßin bir **merkez noktasƒ± (centroid)** belirler ve veri noktalarƒ±nƒ± en yakƒ±n merkeze atar.

K-Means:
- **Classification veya regression yapmaz**
- Veri i√ßindeki **doƒüal gruplarƒ± ke≈üfetmeyi** ama√ßlar
- Segmentasyon problemlerinde sƒ±k√ßa kullanƒ±lƒ±r

---

## üîÑ How does K-Means work? / K-Means Nasƒ±l √áalƒ±≈üƒ±r?

K-Means algoritmasƒ±nƒ±n temel adƒ±mlarƒ±:

- K adet **k√ºme sayƒ±sƒ±** belirlenir.
- Rastgele **K adet centroid** se√ßilir.
- Her veri noktasƒ±, **en yakƒ±n centroid**‚Äôe atanƒ±r.
- Yeni k√ºmelere g√∂re **centroid‚Äôler g√ºncellenir**.
- Centroid‚Äôler sabitlenene kadar i≈ülem **tekrar eder**.

Bu s√ºre√ß sonunda:
- K√ºmeler i√ßi benzerlik **maksimize edilir**
- K√ºmeler arasƒ± fark **maksimize edilir**

---

## üåç Where is K-Means Used? / K-Means Nerelerde Kullanƒ±lƒ±r?

## üß† Why K-Means? / Neden K-Means?

K-Means algoritmasƒ±nƒ±n tercih edilme nedenleri:

- Basit ve anla≈üƒ±lƒ±r bir algoritmadƒ±r.
- B√ºy√ºk veri setlerinde **√ßok hƒ±zlƒ±** √ßalƒ±≈üƒ±r.
- √ñl√ßeklenmi≈ü verilerde olduk√ßa **etkili sonu√ßlar** verir.
- G√∂rselle≈ütirme (2D / 3D) ile kolayca yorumlanabilir.
- Unsupervised learning‚Äôe giri≈ü i√ßin **ideal bir algoritmadƒ±r**.

Ancak:
- K√ºme sayƒ±sƒ± (**K**) √∂nceden belirlenmelidir.
- √ñl√ßeklendirme (scaling) performansƒ± doƒürudan etkiler.
- K√ºresel (spherical) k√ºmelerde daha ba≈üarƒ±lƒ±dƒ±r.

Bu nedenle K-Means, hem **√∂ƒürenme ama√ßlƒ±** hem de **i≈ü d√ºnyasƒ±nda pratik √ß√∂z√ºmler** i√ßin sƒ±k√ßa tercih edilir.

Bu nedenle √∂l√ßeklendirme i≈ülemi zorunludur.


```python
# region Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_processed)

scaled_features[:5]
```

### ‚úÖ 5.3 Adƒ±mƒ± Sonucu / Step 5.3 Outcome

Bu adƒ±mda veri setindeki t√ºm deƒüi≈ükenler **StandardScaler** kullanƒ±larak
ba≈üarƒ±yla √∂l√ßeklendirilmi≈ütir.

- Negatif ve pozitif deƒüerler, g√∂zlemlerin ortalamaya g√∂re

## 6. **K-Means Clustering Model** / _K-Means K√ºmeleme Modeli_

## **6Ô∏è‚É£ K-Means Clustering Model / K-Means K√ºmeleme Modeli**

Bu b√∂l√ºmde, √∂n i≈üleme adƒ±mlarƒ±ndan ge√ßirilen veri seti kullanƒ±larak
**K-Means k√ºmeleme algoritmasƒ±** uygulanacaktƒ±r.

Ama√ß, benzer √∂zelliklere sahip m√º≈üterileri aynƒ± k√ºmeler altƒ±nda toplayarak
anlamlƒ± m√º≈üteri segmentleri olu≈üturmaktƒ±r.

Bu s√ºre√ßte:
- Optimal k√ºme sayƒ±sƒ± belirlenecek
- K-Means modeli eƒüitilecek
- K√ºme etiketleri olu≈üturulacak
- K√ºmeler g√∂rselle≈ütirilip yorumlanacaktƒ±r


## 6.1. üìå **Choosing Optimal K (Elbow Method)** / _Optimal K√ºme Sayƒ±sƒ±nƒ±n Belirlenmesi_

K-Means algoritmasƒ±nda en √∂nemli adƒ±mlardan biri,
k√ºme sayƒ±sƒ±nƒ±n (**K**) doƒüru ≈üekilde belirlenmesidir.

Bu ama√ßla **Elbow Method** kullanƒ±lacaktƒ±r.
Elbow y√∂ntemi, farklƒ± K deƒüerleri i√ßin modelin hata √∂l√ß√ºt√º olan
**WCSS (Within-Cluster Sum of Squares)** deƒüerlerini hesaplar.

WCSS deƒüeri:
- K√ºme sayƒ±sƒ± arttƒ±k√ßa azalƒ±r
- Ancak belirli bir noktadan sonra azalma hƒ±zƒ± yava≈ülar

Bu kƒ±rƒ±lma noktasƒ± (‚Äúdirsek noktasƒ±‚Äù),
optimal k√ºme sayƒ±sƒ±nƒ± temsil eder.

```python
# region Elbow Method

from sklearn.cluster import KMeans

wcss = []

K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(
        n_clusters=k,
        init="k-means++",
        random_state=42
    )
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# region Elbow Plot

plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker="o")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

### üîé Elbow Method Yorumu / Optimal K Analysis

Elbow grafiƒüi incelendiƒüinde:

- **K = 1 ‚Üí 4** aralƒ±ƒüƒ±nda **WCSS deƒüerinde hƒ±zlƒ± bir d√º≈ü√º≈ü** g√∂zlemlenmektedir.
- **K = 4** sonrasƒ±nda WCSS deƒüerindeki azalma belirgin ≈üekilde **yava≈ülamaktadƒ±r**.
- Bu durum, **K = 4** noktasƒ±nda bir ‚Äúdirsek‚Äù olu≈ütuƒüunu g√∂stermektedir.

Bu nedenle, veri seti i√ßin **optimal k√ºme sayƒ±sƒ± K = 4** olarak se√ßilmi≈ütir.
Bu se√ßim, model karma≈üƒ±klƒ±ƒüƒ± ile k√ºme i√ßi homojenlik arasƒ±nda
denge saƒülamaktadƒ±r.

 ## 6.2. ü§ñ **Model Training with K-Means**/ _K-Means Modelinin Eƒüitilmesi_

Bu adƒ±mda, Elbow Method ile belirlenen **optimal k√ºme sayƒ±sƒ± (K = 4)**
kullanƒ±larak K-Means modeli eƒüitilecektir.

Ama√ß, √∂l√ßeklendirilmi≈ü veri seti √ºzerinde
benzer √∂zelliklere sahip g√∂zlemleri aynƒ± k√ºmeler altƒ±nda toplamak
ve her bir g√∂zleme bir **k√ºme etiketi (cluster label)** atamaktƒ±r.

```python
# region K-Means Model Training

kmeans = KMeans(
    n_clusters=4,
    init="k-means++",
    random_state=42
)

cluster_labels = kmeans.fit_predict(scaled_features)

cluster_labels[:10]
```

### ‚úÖ 6.2 Adƒ±mƒ± Sonucu / Step 6.2 Outcome

K-Means modeli, Elbow Method ile belirlenen **K = 4** deƒüeri kullanƒ±larak
ba≈üarƒ±yla eƒüitilmi≈ütir.

- `fit_predict()` fonksiyonu sayesinde model hem eƒüitilmi≈ü
  hem de her g√∂zlem i√ßin bir **k√ºme etiketi (cluster label)** olu≈üturulmu≈ütur.
- √úretilen k√ºme etiketleri **0, 1, 2, 3** aralƒ±ƒüƒ±ndadƒ±r.
- Bu etiketler, her veri noktasƒ±nƒ±n hangi k√ºmeye ait olduƒüunu g√∂stermektedir.

Bu a≈üamada elde edilen k√ºme etiketleri,
bir sonraki adƒ±mda veri setine eklenerek
k√ºmelerin detaylƒ± ≈üekilde analiz edilmesini saƒülayacaktƒ±r.

 ## 6.3. üè∑Ô∏è **Cluster Assignment** / _K√ºme Etiketlerinin Olu≈üturulmasƒ±_


Bu adƒ±mda, 6.2 b√∂l√ºm√ºnde elde edilen **k√ºme etiketleri (cluster_labels)**,
veri setine yeni bir s√ºtun olarak eklenecektir.

Ama√ß:
- Her m√º≈üterinin hangi k√ºmeye ait olduƒüunu veri √ºzerinde g√∂rmek
- K√ºmeleri daha kolay analiz etmek ve g√∂rselle≈ütirmeye hazƒ±rlƒ±k yapmaktƒ±r

Bu i≈ülem sonrasƒ±nda veri seti, k√ºme bazlƒ± inceleme (groupby) i√ßin hazƒ±r hale gelecektir.

```python
# region Cluster Assignment

# K√ºmeleri yorumlamak i√ßin √∂l√ßeklendirme √∂ncesi (okunabilir) veri √ºzerinden ilerliyoruz
df_clustered = df_processed.copy()

# Cluster etiketlerini ekliyoruz
df_clustered["Cluster"] = cluster_labels

df_clustered.head()
```
