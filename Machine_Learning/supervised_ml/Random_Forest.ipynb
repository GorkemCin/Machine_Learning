## **1. üå≥ Introduction** / _Giri≈ü_

## üå≤ **What is Random Forest? / Random Forest Nedir?**

**Random Forest**, birden fazla **Decision Tree (Karar Aƒüacƒ±)** modelinin bir araya gelmesiyle olu≈üturulan bir **ensemble (topluluk)** makine √∂ƒürenimi y√∂ntemidir.  
Hem **sƒ±nƒ±flandƒ±rma (classification)** hem de **regresyon (regression)** problemlerinde kullanƒ±lƒ±r.

Temel mantƒ±ƒüƒ±, bir√ßok karar aƒüacƒ±nƒ±n √ºrettiƒüi sonu√ßlarƒ± birle≈ütirip **daha g√º√ßl√º, daha genel ve daha kararlƒ± (robust)** bir model elde etmektir.

### **Random Forest nasƒ±l √ßalƒ±≈üƒ±r?**

- Model, verinin farklƒ± **√∂rneklem alt k√ºmeleri** √ºzerinde bir√ßok Decision Tree olu≈üturur.
- Her aƒüacƒ±n eƒüitimi sƒ±rasƒ±nda, √∂zelliklerin (features) rastgele bir alt k√ºmesi se√ßilir. Buna **Random Subspace** denir.
- Her aƒüa√ß baƒüƒ±msƒ±z olarak tahmin √ºretir.
- **Classification** ‚Üí √ßoƒüunluk oyu (majority voting)  
- **Regression** ‚Üí ortalama (averaging)

Bu yapƒ± sayesinde Random Forest:

- Decision Tree‚Äôye g√∂re **daha az overfitting** yapar.
- **Genelleme performansƒ±** y√ºksektir.
- G√ºr√ºlt√ºl√º ve karma≈üƒ±k veri setlerinde **√ßok ba≈üarƒ±lƒ±dƒ±r**.
- Non-linear ili≈ükileri √ßok iyi √∂ƒürenir.

Model, hem g√º√ßl√º hem de pratik olduƒüu i√ßin bir√ßok end√ºstride ‚Äúvarsayƒ±lan g√º√ßl√º algoritma‚Äù olarak bilinir.

---

## üåç **Where is Random Forest Used? / Random Forest Nerelerde Kullanƒ±lƒ±r?**

Random Forest, ger√ßek hayatta en √ßok kullanƒ±lan ML algoritmalarƒ±ndan biridir.  
Genelde **y√ºksek doƒüruluk**, **d√º≈ü√ºk overfitting** ve **kolay yorumlanabilirlik** i√ßin tercih edilir.

---

### üß¨ **Classification (Sƒ±nƒ±flandƒ±rma)**

- Hastalƒ±k tahmini (√∂r: kalp hastalƒ±ƒüƒ±, diyabet)
- M√º≈üterinin churn olup olmayacaƒüƒ±nƒ±n tahmini
- Spam e-posta tespiti
- Kredi ba≈üvurusunun onaylanƒ±p onaylanmayacaƒüƒ±
- Sahtekarlƒ±k (fraud) tespiti

---

### üìà **Regression (Regresyon)**

- Ev fiyatlarƒ±nƒ±n tahmini  
- √úr√ºn satƒ±≈ü tahminleri  
- ≈ûirket gelir projeksiyonlarƒ±  
- Enerji t√ºketim tahminleri  

---

### üè¢ **Business Decision Systems (ƒ∞≈ü Karar Sistemleri)**

- Risk analizleri  
- M√º≈üteri segmentasyonu  
- Pazar ara≈ütƒ±rmasƒ± ve talep tahmini  
- Karar destek sistemleri  

---

## üß† **Why Random Forest? / Neden Random Forest?**

Random Forest:

- Tek bir Decision Tree‚Äôye g√∂re **√ßok daha stabil** sonu√ß verir.  
- G√ºr√ºlt√ºye dayanƒ±klƒ±dƒ±r.  
- Feature importance sayesinde **hangi deƒüi≈ükenlerin model i√ßin √∂nemli olduƒüunu g√∂sterir**.  
- Hyperparameter tuning ile √ßok g√º√ßl√º hale getirilebilir.  
- Veri √∂n i≈üleme s√ºreci genelde daha basittir (√∂l√ßeklendirme zorunlu deƒüildir).

Bu nedenle **hem yeni ba≈ülayanlar hem de profesyoneller** tarafƒ±ndan en √ßok tercih edilen algoritmalardan biridir.

## 5.1. üîß **One-Hot Encoding**/ _Kategorik Deƒüi≈ükenlerin D√∂n√º≈üt√ºr√ºlmesi_

Makine √∂ƒürenimi modelleri kategorik verilerle doƒürudan √ßalƒ±≈üamaz.  
Bu y√ºzden kategorik deƒüi≈ükenleri **One-Hot Encoding** y√∂ntemiyle sayƒ±sal forma d√∂n√º≈üt√ºr√ºyoruz.

D√∂n√º≈üt√ºr√ºlecek s√ºtunlar:
- Sex
- ChestPainType
- RestingECG
- ExerciseAngina
- ST_Slope

Bu i≈ülem, her kategori i√ßin ayrƒ± bir s√ºtun olu≈üturarak modelin veriyi daha iyi anlamasƒ±na yardƒ±mcƒ± olur.


```python
# One-hot encoding uygulanacak s√ºtunlar
categorical_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

# One-hot encoding i≈ülemi
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Sonu√ß DataFrame'in ilk 5 satƒ±rƒ±nƒ± g√∂r√ºnt√ºleyelim
df_encoded.head()
```

Bu adƒ±mda modelin tahmin edeceƒüi deƒüi≈ükeni (y) ve tahmin i√ßin kullanƒ±lacak √∂zellikleri (X) birbirinden ayƒ±rƒ±yoruz.

- **Target (y):** HeartDisease  
- **Features (X):** One-hot encoding sonrasƒ± olu≈üan t√ºm s√ºtunlar (HeartDisease hari√ß)


```python
# Target deƒüi≈üken
y = df_encoded["HeartDisease"]

# Features (HeartDisease s√ºtunu hari√ß)
X = df_encoded.drop("HeartDisease", axis=1)

# Kontrol edelim
X.head(), y.head()
```

## 5.3. ‚úÇÔ∏è **Train‚ÄìTest Split**

Bu adƒ±mda veri setini ikiye ayƒ±rƒ±yoruz:

- **Training Set (%80):** Modelin √∂ƒürendiƒüi veri  
- **Test Set (%20):** Modelin hi√ß g√∂rmediƒüi veri (doƒüruluk deƒüerlendirme i√ßin)

Train‚ÄìTest split, modelin ger√ßek hayatta nasƒ±l performans g√∂stereceƒüini √∂l√ßmek i√ßin kritik bir adƒ±mdƒ±r.

```python
from sklearn.model_selection import train_test_split

# %80 eƒüitim, %20 test ≈üeklinde ayƒ±rƒ±yoruz
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y   # sƒ±nƒ±f dengesini korur
)

# Boyutlarƒ± kontrol edelim
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### üìä Train‚ÄìTest Split Sonucu 

Veri seti %80 eƒüitim ve %20 test olacak ≈üekilde ba≈üarƒ±yla ayrƒ±lmƒ±≈ütƒ±r.

- **X_train:** (734, 15)
- **X_test:** (184, 15)
- **y_train:** (734,)
- **y_test:** (184,)

Bu daƒüƒ±lƒ±m ≈üu anlama gelir:

- Model, toplam 734 √∂rnek √ºzerinde **√∂ƒürenme (training)** yapacaktƒ±r.
- Daha √∂nce hi√ß g√∂rmediƒüi 184 √∂rnek ile **test** edilecektir.
- `stratify=y` kullanƒ±ldƒ±ƒüƒ± i√ßin eƒüitim + test setindeki sƒ±nƒ±f daƒüƒ±lƒ±mƒ± korunmu≈ütur.
- Eƒüitim setinin yeterince b√ºy√ºk olmasƒ± Random Forest modelinin daha iyi genelleme yapmasƒ±nƒ± saƒülar.

Bu a≈üama ile, model eƒüitimi i√ßin ideal bir veri hazƒ±rlƒ±ƒüƒ± tamamlanmƒ±≈ü ve bir sonraki adƒ±m olan **Random Forest model kurulumuna** ge√ßmeye hazƒ±r hale gelinmi≈ütir.


## 6. üå≤ **Model Training** / _Modeli Eƒüitme_ - _Random Forest Classifier_ 

Bu b√∂l√ºmde hazƒ±rladƒ±ƒüƒ±mƒ±z eƒüitim verisini kullanarak bir **Random Forest Classifier** modeli eƒüiteceƒüiz.

Random Forest modeli:
- Birden fazla karar aƒüacƒ±nƒ±n birlikte √ßalƒ±≈ütƒ±ƒüƒ± bir topluluk modelidir (ensemble).
- Overfitting'e kar≈üƒ± dayanƒ±klƒ±dƒ±r.
- Karma≈üƒ±k ili≈ükileri iyi √∂ƒürenir.

## 6.1. üå≤ **Random Forest Modelinin Olu≈üturulmasƒ± ve Eƒüitilmesi**

Bu adƒ±mda Random Forest modelini olu≈üturuyor ve eƒüitim verisi √ºzerinde eƒüitiyoruz.  
Model, birden fazla karar aƒüacƒ±nƒ±n birle≈üiminden olu≈üan g√º√ßl√º bir ensemble y√∂ntemidir.

```python
import warnings
warnings.filterwarnings("ignore")
# ‚ÄúBu kod, Python‚Äôdaki t√ºm uyarƒ± mesajlarƒ±nƒ± g√∂rmezden gelerek √ßƒ±ktƒ±larƒ±n daha temiz g√∂r√ºnmesini saƒülar.‚Äù

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Random Forest modelini olu≈üturuyoruz
rf = RandomForestClassifier(
    n_estimators=200,   # 200 aƒüa√ß
    max_depth=None,    # Derinlik sƒ±nƒ±rƒ± yok (model serbest)
    random_state=42    # Tekrarlanabilirlik
)

# 2. Modeli eƒüitiyoruz
rf.fit(X_train, y_train)

# 3. Test verisi ile tahmin yapƒ±yoruz
y_pred = rf.predict(X_test)

# 4. Accuracy deƒüerini alƒ±yoruz
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

bilindiƒüinden, bu deƒüer olduk√ßa ba≈üarƒ±lƒ± kabul edilir.

## 7. üîß **Hyperparameter Tuning (Model Optimization)** / _Hiperparametre Ayarlama (Model Optimizasyonu)_

≈ûu ana kadar varsayƒ±lan ayarlarla g√º√ßl√º bir model kurduk (~%88 accuracy).
Bu adƒ±mda Random Forest‚Äôƒ±n bazƒ± hiperparametrelerini GridSearchCV ile optimize ederek daha iyi genelleme performansƒ± elde etmeyi ve overfitting riskini azaltmayƒ± hedefliyoruz.

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(
    estimator=rf,
    param_grid={
        "n_estimators": [100, 200, 300],
        "max_depth": [None, 10, 20]
    },
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

Bu adƒ±mda Random Forest modeli i√ßin GridSearchCV kullanƒ±larak farklƒ± aƒüa√ß sayƒ±sƒ± (`n_estimators`)
ve maksimum derinlik (`max_depth`) kombinasyonlarƒ± √ßapraz doƒürulama (CV) ile test edilmi≈ütir.
En y√ºksek doƒüruluk (accuracy) deƒüerini veren hiperparametre kombinasyonu se√ßilerek
en iyi model (`best_estimator_`) elde edilmi≈ütir.

## 7.1. üîç **GridSearchCV Results Visualization** / _GridSearchCV Sonu√ßlarƒ±nƒ±n G√∂rselle≈ütirilmesi_

Bu b√∂l√ºmde GridSearchCV‚Äônin denediƒüi parametre kombinasyonlarƒ±nƒ±n ortalama CV accuracy skorlarƒ±nƒ± ƒ±sƒ± haritasƒ± (heatmap) ile g√∂rselle≈ütiriyoruz.

```python
import pandas as pd
import matplotlib.pyplot as plt

# GridSearchCV sonu√ßlarƒ±nƒ± al
results = pd.DataFrame(grid_search.cv_results_)

# Heatmap i√ßin pivot tablo: (max_depth x n_estimators -> mean_test_score)
pivot = results.pivot_table(
    index="param_max_depth",
    columns="param_n_estimators",
    values="mean_test_score"
)

# G√∂rselle≈ütirme
plt.figure(figsize=(10, 6))
plt.imshow(pivot.values, aspect="auto")
plt.colorbar(label="Mean CV Accuracy")
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.xlabel("n_estimators")
plt.ylabel("max_depth")
plt.title("GridSearchCV Heatmap (Mean CV Accuracy)")
plt.show()
```

Isƒ± haritasƒ±, farklƒ± `n_estimators` ve `max_depth` kombinasyonlarƒ±nƒ±n ortalama √ßapraz doƒürulama (CV) doƒüruluk skorlarƒ±nƒ± g√∂stermektedir. 
Daha a√ßƒ±k (y√ºksek) renkler, ilgili hiperparametre kombinasyonlarƒ±nƒ±n modele daha iyi genelleme performansƒ± saƒüladƒ±ƒüƒ±nƒ± ifade eder. 
Bu sonu√ßlara g√∂re √∂zellikle **200‚Äì300 aƒüa√ß sayƒ±sƒ±** ve **daha y√ºksek maksimum derinlik** deƒüerlerinin daha ba≈üarƒ±lƒ± olduƒüu g√∂zlemlenmi≈ütir.


## 7.2. üèÜ **Best Hyperparameters & Optimized Model** / _En ƒ∞yi Hiperparametreler ve Optimize Edilmi≈ü Model_

```python
# Best hyperparameters & best CV score
print("Best Parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)
```

 GridSearchCV, en y√ºksek ortalama CV accuracy veren hiperparametre kombinasyonunu se√ßer.

GridSearchCV sonucunda `max_depth=None` deƒüeri, diƒüer derinlik se√ßeneklerine kƒ±yasla
daha y√ºksek ortalama √ßapraz doƒürulama (CV) doƒüruluƒüu saƒülamƒ±≈ütƒ±r.
Bu durum, Random Forest algoritmasƒ±nƒ±n doƒüasƒ± gereƒüi derin aƒüa√ßlarla √ßalƒ±≈üƒ±rken
overfitting riskini azaltabilmesi ve daha karma≈üƒ±k √∂r√ºnt√ºleri √∂ƒürenebilmesi ile a√ßƒ±klanabilir.

## 7.3. üìä **Optimized Model Evaluation (Test Set Performance)** / _Optimize Edilmi≈ü Modelin Deƒüerlendirilmesi (Test Verisi Performansƒ±)_  

Bu adƒ±mda GridSearchCV ile elde edilen en iyi model, test verisi √ºzerinde
deƒüerlendirilerek ger√ßek genelleme performansƒ± analiz edilmektedir.


```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1) GridSearchCV ile bulunan en iyi modeli alƒ±yoruz
best_rf = grid_search.best_estimator_

# 2) Test seti √ºzerinde tahmin yapƒ±yoruz
y_pred_best = best_rf.predict(X_test)

# 3) Test accuracy
best_test_acc = accuracy_score(y_test, y_pred_best)
print("Optimized Model Test Accuracy:", best_test_acc)

# 4) Classification report
print("\nClassification Report (Optimized Model):\n")
print(classification_report(y_test, y_pred_best))

# 5) Confusion matrix
cm_best = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_best, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (Optimized Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 6) Baseline model ile kar≈üƒ±la≈ütƒ±rma (rf daha √∂nce eƒüitildiyse)
baseline_test_acc = accuracy_score(y_test, y_pred)
print("\nBaseline Model Test Accuracy:", baseline_test_acc)
print("Accuracy Difference (Optimized - Baseline):", best_test_acc - baseline_test_acc)
```

Optimize edilmi≈ü Random Forest modeli test setinde %88.04 accuracy elde etmi≈ütir. Precision, recall ve F1-score deƒüerlerinin her iki sƒ±nƒ±f i√ßin de dengeli olmasƒ±, modelin genelleme performansƒ±nƒ±n g√º√ßl√º ve stabil olduƒüunu g√∂stermektedir.

## 7.4. üîÅ **Model Comparison (Baseline vs Optimized)** / _Model Kar≈üƒ±la≈ütƒ±rmasƒ± (Ba≈ülangƒ±√ß Modeli vs Optimize Edilmi≈ü Model)_


Bu b√∂l√ºmde, GridSearchCV √∂ncesi olu≈üturulan ba≈ülangƒ±√ß (baseline) model ile optimize edilmi≈ü Random Forest modelinin performanslarƒ± kar≈üƒ±la≈ütƒ±rƒ±lmƒ±≈ütƒ±r.

```python
# 7.4 - Baseline vs Optimized Model Comparison

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Baseline model (rf) test tahminleri
y_pred_base = rf.predict(X_test)

# Optimized model (GridSearchCV'den gelen en iyi model) test tahminleri
best_model = grid_search.best_estimator_
y_pred_opt = best_model.predict(X_test)

def metric_summary(y_true, y_pred):
    return {
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1-score": f1_score(y_true, y_pred),
    }

base_metrics = metric_summary(y_test, y_pred_base)
opt_metrics  = metric_summary(y_test, y_pred_opt)

print("üìå Baseline Model (RF) Metrics:")
for k, v in base_metrics.items():
    print(f"- {k}: {v:.4f}")

print("\nüèÜ Optimized Model (GridSearchCV Best Estimator) Metrics:")
for k, v in opt_metrics.items():
    print(f"- {k}: {v:.4f}")

print("\nüìà Difference (Optimized - Baseline):")
for k in base_metrics:
    diff = opt_metrics[k] - base_metrics[k]
    print(f"- {k}: {diff:+.4f}")
```
